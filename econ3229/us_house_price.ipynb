```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

df = pd.read_csv('shiller_home_price_2017.csv')
df=df.drop(df.index[0:63])
df
```

```python
start_date=datetime.date(1953, 1, 1) ## <--- make change: series_start_date
end_date=datetime.date(2017, 8, 1) ## <--- make change: series_end_date

#since we only plot quarterly data, these monthly observations should be transformed in frequency
new_date=pd.date_range(start=start_date, end=end_date, freq='MS')
df['new_d']=new_date
df
```

```python
#this is a third-party api for manipulating fred data
from fred import Fred

#first, use the api key to get access to fred, all data in dictionary format
#replace with your own key
fr = Fred(api_key='4bba000775edea91bc01729d217cf3da', response_type='dict')


nber=fr.series.observations(series_id="USREC")


NBER=pd.DataFrame.from_dict(nber)


NBER.columns = 'new_d', 'end', 'start', 'nber'

NBER
```

```python
comb=pd.merge(df, NBER, how='left', on='new_d')
t=comb[['new_d', 'Index', 'nber']]
t['+nber']=200*t['nber'] 
t['-nber']=-t['+nber'] 
```

```python
#since fill_between mathod doesnt work well with datetype64, we need to transform it
finally_=t.new_d.values
fig, ax=plt.subplots(figsize=(16, 9))
ax.plot(finally_, t['Index'], label='Real home price index', color='red', linewidth=3) ## <--- make change: series_name, series_label


#transform the shaded area into rectangle
import matplotlib.transforms as mtransforms
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.fill_between(finally_, t['+nber'], t['-nber'], facecolor='gray', interpolate=False, transform=trans, alpha=0.2)

ax.set_xlim('1967-08-01', '2017-08-01')

ax.set_ylim(100, 200)
ax.set_yticklabels(np.linspace(100, 300, 11), fontsize=12)
ax.grid()
ax.set_ylabel("%", fontsize=16)
ax.legend(loc='best', fontsize=16)
ax.set_title("U.S. Real Home Prices, 1967-2017", fontsize=20, position=[.5, 1.03])
## <--- make change: graph title
plt.savefig('real_home_price.jpg', dpi=200)
```
