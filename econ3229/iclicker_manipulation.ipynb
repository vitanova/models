```python
"""
I was frustrated by the huge amount of boring and useless work of collecting grades in iclickers in excel.
the problem is four folds:
1) some people don't get his/her name in clikcer the same as in the canvas system
2) their email address also varies
3) 
there are four or five exams and after them I have to repeat the process manually.
so I spend an entire night (11/27/2017) to automize it

Fangda Wang
"""

import pandas as pd
import numpy as np


df=pd.read_csv('iClicker_All_Session_Export.csv')
# step 1
# change name format from 'Wang, Fangda' to 'Fangda Wang'
def change_name(old_name):
    if ', ' not in old_name:
        return old_name
    else:
        return str(old_name.split(', ')[1]+' '+old_name.split(', ')[0]).title()
df['new_name']=df.apply(lambda row: change_name(row['Name']), axis=1)
# drop the last row
df=df[:-1]

# step 2
# get the pawprint for each student
def get_pawprint(email):
    if pd.isnull(email):
        return ' '
    else:
        temp1=str(email.split('@')[0])
        temp2=str(email.split('@')[1])
        if temp2=='mail.missouri.edu' or temp2=='mail.mizzou.edu' or temp2=='missouri.edu' :
            return temp1
        else:
            return ' '
df['pawprint']=df.apply(lambda row: get_pawprint(row['Email']), axis=1)

# step 3
# calculate the total score for each session and attendance
for j in range(1, 42):
    keyword='Session '+str(j)+' '
    Ses=[]
    for i in list(df):
        if keyword in i:
            Ses.append(i)    
    if np.shape(Ses)[0] is not 0:
        df[keyword]=df[Ses[0]]+df[Ses[1]]

def attend(row):
    cols=[s for s in list(df) if '- Participation' in s]
    temp=[]
    for col in cols:
        temp.append(row[col])
    return sum(temp)*2

df['attendence']=df.apply(attend, axis=1)

# step 4
# change scores for those students
# I introduce an copy of df to make sure the score changing process will not be multiply executed
df2 = df.copy()
def change_score(nam, dat, delta):
    ind=df2[df2.new_name==nam].index[0]
    ses=[s for s in list(df2) if dat in s][0].split(dat)[0]
    df2.set_value(ind, ses, df.at[ind, ses]+delta)

# Ryan Seidel +2 on 09/08
change_score('Ryan Seidel', '- 09/08', 2)
# Jackson Schroeder +1 on 09/11
change_score('Jackson Schroeder', '- 09/11', 1)
# Christina Niedbalski +2 on 10/11
change_score('Christina Niedbalski', '- 10/11', 2)
# Riley Miller+2.5 on 10/16
change_score('Riley Miller', '- 10/16', 2.5)
# Trevor Wojcik +2.5 on 10/16
change_score('Trevor Wojcik', '- 10/16', 2.5)
# Lotanna Ofokansi +2 on 10/20
change_score('Lotanna Ofokansi', '- 10/20', 2)
# James Thompson +2 on 10/23
change_score('James Thompson', '- 10/23', 2)
# Shayne Bass +1 on 10/30
change_score('Shayne Bass', '- 10/30', 1)
# Killian Kamp +1 on 10/30
change_score('Killian Kamp', '- 10/30', 1)

df2
```

```python
# step 5
# all attendence on 10/16 get full credit (2.5p)
df3=df2.copy()
participate=[s for s in list(df3) if '- 10/16' in s][1]
ses=[s for s in list(df3) if '- 10/16' in s][0].split('- 10/16')[0]
def update_1016(row):
    if row[participate]>0:
        return 2.5
    else:
        return 0.0
df3[ses]=df3.apply(lambda row: update_1016(row), axis=1)

# step 6
# get the best 25 scores
df4=df3.loc[:, 'Session 6 ': 'Session 41 ']
top25 = list()
def process(row):
    top25.append(row.sort_values(ascending=False).head(25))
df4.apply(process, axis=1)
result=[]
for i in range(240):
    result.append(sum(top25[i]))

df3['bo25']=result
df5=df3.loc[:, ['Name', 'Email', 'new_name', 'pawprint', 'bo25', 'attendence']]

# set the score of Connor McAteer
# note that I used the title() method, which results in 'Connor Mcateer'
ind0=df5[df5.new_name=='Connor Mcateer'].index[0]
ind1=df5[df5.new_name=='9EC5316A'].index[0]
df5.set_value(ind0, 'bo25', df5.at[ind1, 'bo25'])
df5=df5.drop([ind1])
df5
```

```python
df5.to_csv('df55.csv')
```

```python
# step 7
# merge and clean data
name_to=pd.read_csv('name_list.csv')
name_to.columns = ['new_name', 'ID', 'pawprint']
name_to=name_to.loc[:, ['new_name', 'pawprint']]
name_to=name_to[:-1]

final1=pd.merge(name_to, df5, how='left', on='pawprint')
final1.rename(columns={'new_name_x': 'new_name'}, inplace=True)
final2=pd.merge(name_to, df5, how='left', on='new_name')
final3=pd.merge(final1, final2, how='left', on='new_name')

def final_score(row):
    if pd.isnull(row['bo25_x']):
        if pd.isnull(row['bo25_y']):
            return 'NaN'
        else:
            return row['bo25_y']
    else:
        return row['bo25_x']
final3['true_25']=final3.apply(final_score, axis=1)

def final_atten(row):
    if pd.isnull(row['attendence_x']):
        if pd.isnull(row['attendence_y']):
            return 'NaN'
        else:
            return row['attendence_y']
    else:
        return row['attendence_x']
final3['true_atten']=final3.apply(final_atten, axis=1)
final3=final3.loc[:, ['new_name', 'pawprint', 'Email_x', 'true_atten', 'true_25']]
final3=final3.drop_duplicates(subset=['new_name'], keep='first')
final3
```
