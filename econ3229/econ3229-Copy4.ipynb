```python
"""
Code for accommodating the graph-making need of ECON 3229-2 Money, Banking, and Financial Systems 

University of Missouri-Columbia

Fangda Wang

Fall 2017


https://github.com/vitanova
"""

# This file is aimed to demonstate the usefulness of the code I wrote before
# I changed several parameters in 1 min or 2, and a new graph is plotted as required

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

#this is a third-party api for manipulating fred data
from fred import Fred

#first, use the api key to get access to fred, all data in dictionary format
fr = Fred(api_key='abcdefghijklmnopqrstuvwxyz123456', response_type='dict') ## <--- replacing with your own key

#now, use the series id to get observations
Y10=fr.series.observations(series_id="GS10") ## <--- make change: series_id
Y025=fr.series.observations(series_id="TB3MS") ## <--- make change: ibid
Y1=fr.series.observations(series_id="GS1") ## <--- make change: ibid
Y2=fr.series.observations(series_id="GS2") ## <--- make change: ibid
Y5=fr.series.observations(series_id="GS5") ## <--- make change: ibid
Y20=fr.series.observations(series_id="GS20") 
Y30=fr.series.observations(series_id="GS30") 

#convert the dictionary data to data frame
y10=pd.DataFrame.from_dict(Y10)
y025=pd.DataFrame.from_dict(Y025)
y1=pd.DataFrame.from_dict(Y1)
y2=pd.DataFrame.from_dict(Y2)
y5=pd.DataFrame.from_dict(Y5)
y20=pd.DataFrame.from_dict(Y20)
y30=pd.DataFrame.from_dict(Y30)

y025.columns = 'date', 'end', 'start', 'tb3ms'
y1.columns = 'date', 'end', 'start', 'gs1'
y2.columns = 'date', 'end', 'start', 'gs2'
y5.columns = 'date', 'end', 'start', 'gs5'
y10.columns = 'date', 'end', 'start', 'gs10'
y20.columns = 'date', 'end', 'start', 'gs20'
y30.columns = 'date', 'end', 'start', 'gs30'


#merge the two tables and use date as key
al=pd.merge(y025, y1, how='left', on='date')
al=pd.merge(al, y2, how='left', on='date')
al=pd.merge(al, y5, how='left', on='date')
al=pd.merge(al, y10, how='left', on='date')
al=pd.merge(al, y20, how='left', on='date')
al=pd.merge(al, y30, how='left', on='date')
al=al.drop('end_x', axis=1)
al=al.drop('start_x', axis=1)
al=al.drop('end_y', axis=1)
al=al.drop('start_y', axis=1)
al=al.drop('end', axis=1)
al=al.drop('start', axis=1)
al.head()
```
