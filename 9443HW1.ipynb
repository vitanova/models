```python
import pandas as pd
import numpy as np
```

```python
m1 = pd.read_csv('C:\\Users\\vitan\\Desktop\\18fall\\9443\\model1.csv')
m2 = pd.read_csv('C:\\Users\\vitan\\Desktop\\18fall\\9443\\model2.csv')
m3 = pd.read_csv('C:\\Users\\vitan\\Desktop\\18fall\\9443\\model3.csv')
m4 = pd.read_csv('C:\\Users\\vitan\\Desktop\\18fall\\9443\\model4.csv')
```

```python
m4
```

```python
variables = list(m4.BYGRADS)
names = ['Coef.', 'Std. Err.']
iterables = [variables, names]
ind = pd.MultiIndex.from_product(iterables, names=['var', 'val'])
df2 = pd.DataFrame(index=ind)
df3 = df2.reset_index()
#new var name
newvar = [df3.loc[i,'var'] for i in range(df3.shape[0])]
for i in range(df3.shape[0]):
    if i%2 == 1:
        newvar[i] = np.nan
df3['Variables'] = newvar
def rep(var):
    try:
        return var.replace('_', '\_')
    except:
        return np.nan
df3['Variables'] = df3.apply(lambda row: rep(row['Variables']), axis=1)
df3
```

```python
def find_it(var, val, model):
    try:
        temp = model.loc[model.BYGRADS==var][val].values[0]
        if val == 'Coef.':
            temp1 = model.loc[model.BYGRADS==var]['Std. Err.'].values[0]
            if temp1 == '(omitted)':
                return '0.0000'
            else:
                stderr = float(temp1)
                t_stat = temp/stderr
                if abs(t_stat) > 2.576:
                    return '{0:.4f}'.format(temp) + '***'
                elif abs(t_stat) > 1.960:
                    return '{0:.4f}'.format(temp) + '**'
                elif abs(t_stat) > 1.645:
                    return '{0:.4f}'.format(temp) + '*'
                else:
                    return '{0:.4f}'.format(temp)
        else:
            if temp == '(omitted)':
                return temp
            else:
                return '({0:.4f})'.format(float(temp))
    except:
        return np.nan
df3['(1)'] = df3.apply(lambda row: find_it(row['var'], row['val'], m1), axis=1)
df3['(2)'] = df3.apply(lambda row: find_it(row['var'], row['val'], m2), axis=1)
df3['(3)'] = df3.apply(lambda row: find_it(row['var'], row['val'], m3), axis=1)
df3['(4)'] = df3.apply(lambda row: find_it(row['var'], row['val'], m4), axis=1)
df4 = df3.drop(columns=['var','val'])
df4
```

```python
df4.to_csv('9443HW1.csv',encoding='utf-8', index=None)
```
