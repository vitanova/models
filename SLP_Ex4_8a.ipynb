```python
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure(figsize=(12, 10))
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(0, 5, 0.1)
Y = np.arange(0, 5, 0.1)
X, Y = np.meshgrid(X, Y)
R = (X*Y)**0.5
Z =R.copy()
for (i, j), r in np.ndenumerate(R):
    if i<j:
        Z[i, j] = 0
    else:
        Z[i, j] = r
        
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

# Plot the surface
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)
ax.plot_surface(X,
                Y,
                Z,
                rstride=1, cstride=1,
                cmap=cm.jet,
                alpha=0.5,
                linewidth=0.1)
ax.view_init(ax.elev, 255)

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

ax.set_title('SLP Exercise 4.8 a', fontsize=20)
label = r"\begin{eqnarray*} f(x) &= x_1^{1/2}x_2^{1/2},\  if \ x_2 \geq x_1 \\ &= 0, \ \ \ \ \ \ otherwise \ \ \ \ \ \end{eqnarray*}"
ax.text(0, 5, 4, label, fontsize=20)
plt.tight_layout()
plt.show()
#plt.savefig('SLP Exercise 4_8a', dpi=250)
```
