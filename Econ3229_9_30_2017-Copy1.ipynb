```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

df = pd.read_csv('111.csv')
df.tail(10)
```

```python
aa=[1927.03]
for i in range(1, 182):
    if int(repr(aa[i-1])[-1])==3:
        temp=aa[i-1]+0.06
        aa.append(temp)
    else:
        temp=aa[i-1]+1-0.06
        aa.append(temp)

bb=np.empty(182)
d = {'Date' : aa, 'value' : bb}
new=pd.DataFrame(d, index=aa)
new_one=pd.merge(new, df, how='left', on='Date')
new_one
```

```python
new_one['shift_20']=new_one.P.shift(40)
new_one['shift_10']=new_one.P.shift(20)
new_one['shift_3']=new_one.P.shift(6)

def back_20(x, y):
    return ((x/y)**0.05-1)*100
def back_10(x, y):
    return ((x/y)**0.1-1)*100
def back_3(x, y):
    return ((x/y)**0.333-1)*100
new_one['back_20']=new_one.apply(lambda row: back_20(row['P'], row['shift_20']), axis=1)
new_one['back_10']=new_one.apply(lambda row: back_10(row['P'], row['shift_10']), axis=1)
new_one['back_3']=new_one.apply(lambda row: back_3(row['P'], row['shift_3']), axis=1)

new_one.tail(10)
```

```python
def D_P(x, y):
    return x/y*100
new_one['D_P']=new_one.apply(lambda row: D_P(row['D'], row['P']), axis=1)


DPP=np.empty(182)
for i in range(182):
    DPP[i]=new_one.D_P[i]

D_P20=np.empty(182)
D_P10=np.empty(182)
D_P3=np.empty(182)
crit_index=int(new_one.Date[new_one.Date>1947.02].index.tolist()[0])
for i in range(182):
    if i< crit_index:
        D_P20[i]=np.nan
    else:
        D_P20[i]=np.average(DPP[i-40: i])
        

for i in range(182):
    if i< crit_index:
        D_P10[i]=np.nan
    else:
        D_P10[i]=np.average(DPP[i-20: i])

for i in range(182):
    if i< crit_index:
        D_P3[i]=np.nan
    else:
        D_P3[i]=np.average(DPP[i-6: i])

new_one['D_P20']=D_P20
new_one['D_P10']=D_P10
new_one['D_P3']=D_P3

def Total(x, y):
    return x+y
new_one['Total20']=new_one.apply(lambda row: Total(row['back_20'], row['D_P20']), axis=1)
new_one['Total10']=new_one.apply(lambda row: Total(row['back_10'], row['D_P10']), axis=1)
new_one['Total3']=new_one.apply(lambda row: Total(row['back_3'], row['D_P3']), axis=1)
zzz=new_one.drop(new_one.index[0:40])
zzz.head()
```

```python
aaa=np.linspace(1947, 2017.5, 142)
bbb=np.zeros(142)
fig, ax =plt.subplots(figsize=(13, 8))
ax.plot(aaa, zzz['back_20'], color='red', lw=2, label='20-Year S&P 500')
ax.plot(aaa, zzz['D_P20'], color='blue', lw=2, label='Dividend Yield', linestyle='--')
ax.plot(aaa, zzz['Total20'], color='purple', lw=2, label='Avg Total Return')
ax.plot(aaa, bbb, color='black', label='_nolegend_')
label_2=np.linspace(1947, 2017, 15)
label_3=np.linspace(-4, 18, 12)
label_33=[]
for i in range(12):
    label_33.append(str(label_3[i])+'%')
ax.set_yticks(label_3)
ax.set_yticklabels(label_33)
ax.set_ylim(-4, 18)
ax.set_xlim(1947, 2017.5)
ax.set_xticks(label_2)
ax.text(1982, 13, 'Avg Total Return', fontsize=15, color='purple')
ax.text(1994, 7, '20-Year S&P 500', fontsize=15, color='red')
ax.text(2006, 1, 'Dividend Yield', fontsize=15, color='blue')
ax.set_title('20-Year Moving Average S&P 500 Returns', fontsize=20, position=[.5, 1.03])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.patch.set_facecolor('grey')
ax.patch.set_alpha(0.06)
ax.yaxis.grid()
plt.show()
plt.savefig('20_year.jpg', dpi=300, facecolor=fig.get_facecolor(), edgecolor='none')
```
